## Check how many times each restaurant name appears

```python
# Count how many times each restaurant name appears
name_counts = data['Restaurant Name'].value_counts()

# Filter only those with more than 1 occurrence (i.e., potential chains)
restaurant_chains = name_counts[name_counts > 1]
print("Potential restaurant chains found:")
print(restaurant_chains)
```

## Filter the main data for only chain restaurants

```python
# Filter original data for only restaurants that are part of chains
chains_data = data[data['Restaurant Name'].isin(restaurant_chains.index)]
```

## Analyze average rating and popularity (votes) per chain

```python
# Group by restaurant name and calculate average rating and votes
chain_summary = chains_data.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'mean',
    'City': 'count'  # to show in how many locations the chain exists
}).rename(columns={
    'Aggregate rating': 'Avg Rating',
    'Votes': 'Avg Votes',
    'City': 'Branch Count'
}).round(2)

# Sort by popularity or rating
chain_summary_sorted = chain_summary.sort_values(by='Avg Votes', ascending=False)

print("Top restaurant chains by average votes:")
print(chain_summary_sorted.head(10))
```

## Visualization — Compare Chains

```python
# Bar chart of top chains by rating
top_chains = chain_summary_sorted.head(10)

top_chains[['Avg Rating', 'Avg Votes']].plot(kind='bar', figsize=(8,6), secondary_y='Avg Votes', colormap='Set2')

plt.title('Top Restaurant Chains: Average Rating and Votes')
plt.ylabel('Average Rating')
plt.xlabel('Restaurant Chain')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
